```python
#syntax3 = strobj[ :END]
#here, we specifying the END index not the begin index 
#if we don't specify BEGIN index then pvm takes from first character to END-1 index
a="python"
print(a[:3])
print(a[:6])
print(a[:23])#Index 23 is beyond length. Python safely returns till end.
print(a[:-12]) #-12 means 12 positions from end. But string length is only 6. So Python adjusts it before start, result becomes empty.
print(a[:-1])
```

```python
#syntax4 = strobj[:]
#here we are not specifying both BEGIN & END index 
#if we don't specify the both indexes pvm takes BEGIN index as 0 or -len(strobj) and END index as len(strobj) or -1
#hence this syntax always gives total string data
a="python"
print(a[:])
print(a[0:])
print(a[:len(a)])
print(a[-len(a):])
```

```python
#syntax5 = strobj[BEGIN:END:STEP]
#RULE1 = here, the begin,end & step values can either be +ve/-ve
#RULE2 = If step is positive â†’ Python slices forward. It takes characters from begin to end-1. Condition: begin < end, otherwise result is empty ('').
a="python"
print(a[0:4])
print(a[0:4:1])
print(a[0:6:2])
print(a[2:4:3])
print(a[3:6:1])
print(a[-6:-1:2])
print(a[-6::])
print(a[:6:])
print(a[0:6:])
print(a[:-1:])
```
